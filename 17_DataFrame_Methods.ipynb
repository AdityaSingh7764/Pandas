```python
points_def =  pd.DataFrame(
    {
        '1st Point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd Point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_def
```

```python
def euclidean(row):
  pt_A = row['1st Point']
  pt_B = row['2nd Point']

  return ((pt_B[0] - pt_A[0])**2 + (pt_B[1] - pt_A[1])**2)**0.5
```

```python
points_def['distance'] = points_def.apply(euclidean,axis=1)         # 1 for row
points_def
```
