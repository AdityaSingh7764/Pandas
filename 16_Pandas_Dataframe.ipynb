# **1. What is Dataframe**

# **4. Maths Methods**

```python
# sum (axis argument)(applies sum to every column)

# print(movies.sum())  # can only concatenate string
print(students,'\n')
print(students.sum(axis=0),'\n') # 0 = sum column
print(students.sum(axis=1),'\n')      # 1 = sum row
print(students.mean(axis=0),'\n')
print(students.max(axis=0),'\n')
print(students.median(axis=1))
```

# **6. Selecting Rows from a DataFrame**
* iloc - searches using index positions
* loc - searches using index labels

```python
student_dict = {
    'name':['Sakshi','Riya','Sneha','Akash','Amit','Megha'],
    'iq':[100,90,120,80,0,0],
    'marks':[80,70,100,50,0,0],
    'package':[10,7,14,2,0,0]
}

students = pd.DataFrame(student_dict)
students.set_index('name',inplace=True)
print(students)
```

```python
# single row

print(movies,'\n')
print(movies.iloc[0],'\n')   # iloc works on the bases of items index
print(movies.iloc[3],'\n')
print(type(movies.iloc[0]))
```

```python
# multiple rows
# iloc

print(movies.iloc[4:8],'\n')
print(movies.iloc[4:8:2],'\n')
print(movies.iloc[:3])
```

```python
# fancy indexing

movies.iloc[[0,4,7]]
```

```python
# loc

print(students)
print(students.loc['Sneha'],'\n')
print(students.loc['Sakshi':'Akash'],'\n')
print(students.loc['Sakshi':'Akash':2],'\n')
print(students.loc[['Sakshi','Akash','Megha']])
```

```python
print(students.iloc[0],'\n')
print(students.iloc[0:4],'\n')
print(students.iloc[[0,2,4]])
```

# **8. Filtering a DataFrame**

```python
ipl.head(2)
```

```python
# find all the final winners

final = ipl['MatchNumber'] == 'Final'
df = ipl[final]
print(df[['Season','WinningTeam']],'\n')

# in one line

print(ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']])
```

```python
# how many super over finishes have occured

ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches csk won in kolkata

ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# toss winner is match winner in percentage

(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0] / ipl.shape[0])*100
```

```python
movies.head(2)
```

```python
# movies with rating higher than 8 and votes > 10000

movies[(movies['imdb_rating'] > 8) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
# Action movies with rating higher than 7.5

mask1 = movies['genres'].str.split('|').apply(lambda x:'Action' in x)
# mask1 = movies['genres'].str.contains('Action')
mask2 = movies['imdb_rating'] > 7.5
movies[mask1 & mask2]
```
