# **What is Pandas ?**
Pandas is a fast, powerful, felixble and easy to use open source data analysis and manipulation tool, built on top of the Python programming languages.

# **Pandas Series**
A Pandas Series is like a column in a table. It is 1-D array holding data of any type.

```python
# Importing Pandas

import numpy as np
import pandas as pd
```

# **1. Series from Lists**

```python
# string

fruits = ['Apple','Banana','Pear','Kiwi','Grapes']

print(pd.Series(fruits))
```

```python
# integers

primes = [2,3,5,7,11,13]

print(pd.Series(primes))
```

```python
# Also can give custom index

marks = [84,43,62,55]
subjects = ['computer_science','physics','chemistry','maths']

print(pd.Series(marks,subjects))
```

```python
# Setting a name

marks = [84,43,62,55]
subjects = ['computer_science','physics','chemistry','maths']
print(pd.Series(marks,index=subjects,name="Vivek's Marks"),'\n\n')
Vivek = pd.Series(marks,index=subjects,name="Vivek's Marks")
Vivek
```

# **2. Series from Dictionaries**

```python
marks= {
    'computer_science':84,
    'physics':43,
    'chemistry':62,
    'maths':55
}

print(pd.Series(marks))
marks_series = pd.Series(marks,name="Vivek's Marks")
marks_series
```

# **3. Series Attributes**

```python
# dtype

print(marks_series.dtype)
```

```python
# name

print(marks_series.name)
```

```python
# is_unique

print(marks_series.is_unique,'\n')
marks_series = pd.Series([84,84,43,62,55])
print(marks_series.is_unique)
```

```python
# index

marks_series = pd.Series(marks,subjects)
print(marks_series.index)
```

```python
# values

print(marks_series.values)
```

# **5. Methods of Series**

```python
# head and tail
# print first and last 5 rows in the dataset (default)
temp_subs = pd.read_csv('/content/subs.csv')
subs = temp_subs.squeeze(True)
print(subs,'\n')

print(subs.head(),'\n')
print(subs.tail(),'\n')
print(subs.head(8))
```

```python
# sample
temp_movies = pd.read_csv('/content/bollywood.csv',index_col='movie')
movies = temp_movies.squeeze(True)
print(movies,'\n')

print(movies.sample(),'\n')            # randomly selects movies
print(movies.sample(4))
```

```python
# value_counts
print(movies,'\n')

print(movies.value_counts())
```

```python
# sort_values  (do temporary changes)
print(movies,'\n')

print(movies.sort_values(),'\n')
print(movies.sort_values(ascending=False),'\n')   # to sort in descending order
```

```python
temp_score = pd.read_csv('/content/kohli_ipl.csv',index_col='match_no')
score = temp_score.squeeze(True)
print(score.sort_values(ascending=False).head(1),'\n')
print(score.sort_values(ascending=False).head(1).values)  # method chaining
```

```python
# print(score.sort_values(inplace=True),'\n')   # do permanent sorting to the dataset
```

# **6. Maths Methods in Series**

```python
# count  (doesn't count missing values)

print(score.count())
```

```python
# sum, product

print(subs.sum())
```

```python
# mean, median, mode, std, var

print(subs.mean(),'\n')
print(score.mean(),'\n')
print(movies.mode(),'\n')
print(subs.std(),'\n')
print(subs.var())
```

```python
# min, max

print(subs.min())
print(subs.min())
```

```python
# describe

print(subs.describe())
```

# **7. Series Indexing**

```python
# integer indexing

a = pd.Series([12,35,64,23,6,78,45,3,14])
print(a[7])
```

```python
# negative indexing doesn't work in Series

# print(a[-1])
```

```python
print(movies,'\n')
print(movies[0],'\n')
print(movies['Why Cheat India'],'\n')
print(movies[-1],'\n')
```

```python
print(marks_series,'\n')          # negative indexing only works with string datatype
print(marks_series[-1])
```

```python
# positive slicing

print(score,'\n')
print(score[5:16])
```

```python
# negative silcing

print(score,'\n')
print(score[-5:],'\n')
print(movies,'\n')
print(movies[-5:])
```

```python
# fancy indexing

print(score[[1,3,4,5]])
```

```python
# indexing with labels - fancy indexing

print(movies['Saathiya (film)'])
```

# **8. Editing Series**

```python
print(marks_series,'\n')
marks_series[2] = 86
print(marks_series)
```

```python
# if index doesn't not exist

print(marks_series,'\n')
marks_series['political_science']=75        # It will not throw an error, add it as new element
print(marks_series)
```

```python
# slicing

print(score,'\n')
score[0:2]=[100,90]
print(score)
```

```python
# fancy indexing

score[[1,2,4]]=[100,100,100]
print(score,'\n')
```

```python
# using index label

print(movies,'\n')
movies['Uri: The Surgical Strike'] = 'Ajay Devgn'
print(movies)
```

# **9. Series with Python Functionalities**

```python
# len/type/dir/sorted/min/max

print(subs,'\n')
print(len(subs),'\n')
print(type(subs),'\n')
print(dir(subs),'\n')
print(sorted(subs),'\n')      # gives output in list
print(min(subs),'\n')
print(max(subs))
```

```python
# type conversion

print(type(marks_series))
print(list(marks_series),'\n')
print(dict(marks_series))
```

```python
# membership operator

print(movies,'\n')
print('1920: London' in movies,'\n')   # will only works on index
print('Aftab Shivdasani' in movies.values)
```

```python
# Loops

for i in movies:           # works on values
  print(i)
```

```python
for i in movies.index:    # print movies
  print(i)
```

```python
# Arithmetic Operators

print(marks_series,'\n')
print(100 - marks_series)        # Broadcasting
```

```python
# Relational Operators

print(score,'\n')
print(score>50)
```

# **10. Boolean Indexing on Series**

```python
# Find the no. of 50s and 100s in score dataset

print(score[score>50],'\n')
print(score[score>50].size)
```

```python
# find the ducks

print(score[score==0].size)
```

```python
# count the number of day when I had more than 200 subs on a day

print(subs[subs>200].size)
```

```python
# find the actors who have done more than 20 movies

num_movies = movies.value_counts()
print(num_movies,'\n')
print(num_movies[num_movies>20],'\n')
print(num_movies[num_movies>20].size)
```
